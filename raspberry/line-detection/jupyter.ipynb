```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path = os.path.expanduser("~/Code/line-following-robot/docs/images-dataset/black-curve.png")
```

```python
img_grey = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

plt.imshow(img_grey, cmap = "gray")
```

```python
_, mask = cv2.threshold(img_grey, 50, 255, cv2.THRESH_BINARY_INV)

plt.imshow(mask, cmap="gray")
```

```python
# normalisation and making left side negative

mask_norm = mask.astype(float) / 255

width = len(mask_norm[0])
signed = mask_norm
signed[:, :width//2] *= -1

plt.imshow(signed, cmap="RdBu", vmin=-1, vmax = 1)
plt.colorbar()
plt.axvline(x=width//2, color='green', linewidth=2)
```

```python
# collapce columns and get the sum for each row

sum = signed.sum(axis=1) / 255

plt.plot(sum)
```

```python
weights = np.linspace(start=0, stop=1, num=len(sum))
weighted_sum = weights * sum

position = weighted_sum.sum()

print(position)
```
